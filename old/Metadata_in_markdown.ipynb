### **Summary**
- **Content:**
  - **Trips:** Average number of trips per person per day/year.
  - **Distance Travelled:** Average distance travelled per person per day/year.

## Definitions and Explanations

### **General Definitions**
- **Trip:**
  - A one-way course of travel with a single main purpose.
  - **Example:** Home to work is one trip, regardless of transport modes used.
  
- **Stage:**
  - A segment of a trip involving a specific mode of transport or vehicle change.

- **Regular Trips:**
  - Daily mobility of residents aged 6+ in private households.
  - Includes domestic holiday mobility and certain work-related movements.
  - Excludes series of calls trips and professional heavy freight transport.

- **Regular Passenger Kilometres:**
  - Kilometers traveled during regular trips within Dutch territory.
  - Calculated based on the stage mode of transport.

- **Series of Calls Trips:**
  - Continuous trips with 4+ consecutive stops for the same main purpose related to work.
  - **Examples:** Taxi drivers, parcel deliverers, doctor’s rounds.

- **Domestic Holiday Mobility:**
  - Trips due to holidays within the Netherlands or travel on Dutch territory during holidays abroad.
  - Includes trips for relaxation or fun, excluding family or acquaintance visits at home.

### **Symbols Explanation**
- **Empty cell:** Not applicable
- **`.` (Dot):** Unknown, unreliable, or confidential figure
- **`*` (Asterisk):** Provisional figure
- **`**` (Double Asterisk):** Revised provisional figure
